```python
del date
from datetime import datetime, date ,timedelta
```

```python
df_noncovid.iloc[1][5]
```

```python
for i in range(0,1089): #전체 열을 대상으로 반복문을 통해 column 분리
    for j in range (1,(int(df_noncovid.loc[i,'DK_DAY']+1))): #해당 열에 해당하는 것 column 분리
        if (int(df_noncovid.loc[i,'DK_DAY'])==0):
            break
        ymd = str(df_noncovid.iloc[i][5])
        year = int((ymd)[0:4])
        month = int((ymd)[4:6])
        day = int((ymd)[6:8])
        temp = date(year, month, day)+timedelta(days=j)
        string = str(temp.year)
        if (temp.month<10):
            string = string + str(0)
            string = string + str(temp.month)
        else:
            string = string + str(temp.month)
        if (temp.day<10):
            string = string + str(0)
            string = string + str(temp.day)
        else:
            string = string + str(temp.day)
        
        Dict = {'DK_SS':[int(df_noncovid.iloc[i][0])],
                'DK_NAME':[str(df_noncovid.iloc[i][1])],
                'EVENT_GUBUN':[str(df_noncovid.iloc[i][3])],
                'EVENT_NAME':[str(df_noncovid.iloc[i][4])],
                'RENT_START_YMD':[string],
                'JS_PERSON':[df_noncovid.iloc[i][9]]}
        myDF = pd.DataFrame(Dict)
        df_fix = pd.concat([df_fix,myDF],ignore_index=True)#해당 날짜를 모두 넣어야 함.
```

## 요일별 데이터 분석을 위한 작업

```python
def what_day_is_it(date):
    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    day = date.weekday()
    return days[day]
```

```python
str(df_fix.loc[3,'RENT_START_YMD'])[6:8]
```

```python
for i in df_fix['RENT_START_YMD']:
    ymd = str(i)
    year = int((ymd)[0:4])
    month = int((ymd)[4:6])
    day = int((ymd)[6:8])
```

```python
day_list=[]

for i in df_fix['RENT_START_YMD']:
    ymd = str(i)
    year = int((ymd)[0:4])
    month = int((ymd)[4:6])
    day = int((ymd)[6:8])
    string = what_day_is_it(date(year, month, day))
    day_list.append(string)
```

```python
df_fix['end_day'] = day_list
```
