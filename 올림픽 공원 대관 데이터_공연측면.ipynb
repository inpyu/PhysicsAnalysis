## 요일별 데이터 분석을 위한 작업

```python
def what_day_is_it(date):
    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    day = date.weekday()
    return days[day]
```

```python
day_list=[]

for i in df_event['RENT_START_YMD']:
    ymd = str(i)
    year = int((ymd)[0:4])
    month = int((ymd)[4:6])
    day = int((ymd)[6:8])
    string = what_day_is_it(date(year, month, day))
    day_list.append(string)
```

```python
df_event['start_day'] = day_list
```

```python
df_event['start_day'].unique()
```

```python
df_event
```

```python
day_list=[]

for i in df_event['RENT_END_YMD']:
    ymd = str(i)
    year = int((ymd)[0:4])
    month = int((ymd)[4:6])
    day = int((ymd)[6:8])
    string = what_day_is_it(date(year, month, day))
    day_list.append(string)
```

```python
df_event['end_day'] = day_list
```
