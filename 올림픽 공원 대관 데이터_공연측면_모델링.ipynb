```python
df_trans = pd.read_csv('data.csv')
```

```python
df_trans.drop(columns='Unnamed: 0',inplace=True)
```

```python
df_trans
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_trans.loc[:,df_trans.columns!='티켓 판매수'],df_trans['티켓 판매수'],random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
scores = cross_val_score(lr, X_train, y_train)
print("Cross-validation score:", scores) #train_set에 대한 cross validation 진행
print("Average cross-validation score: {:.2f}".format(scores.mean())) #test에 대한 결과 평균
scores = cross_val_score(lr, X_test, y_test) # train에 대한 결과 평균
print("Cross-validation score:", scores) #train_set에 대한 cross validation 진행
print("Average cross-validation score: {:.2f}".format(scores.mean())) #test에 대한 결과 평균
```

오히려 Test Set에서 0.6까지 보이는 것을 보인다, 하지만 -값을 띄는 것을 보아 좋지 않은 방식의 모델임을 파악할 수 있다. 또한, 음수부터 0.6까지의 데이터로 폭넓은 정확도를 보이는 것을 보아 일반화가 잘 되지도 않은 모델임을 추측할 수 있다.

```python
param_grid = {'alpha': [0.01,0.1,1,10,100]} 
```

```python
grid_search = GridSearchCV(Ridge(), param_grid, cv=5, return_train_score=True)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Optimal Parameters:", grid_search.best_params_)
print("Best Cross Validation Score: {:.2f}".format(grid_search.best_score_))
```

```python
def predictors_plot(model):
    predictors = X_train.columns
    coef = pd.Series(model.coef_,predictors).sort_values()
    coef.plot(kind='bar',title='Model Coefficients')
```

```python
ridge = Ridge(alpha = 100)
```

```python
ridge.fit(X_train,y_train)
```

```python
predictors_plot(ridge)
```

## Decision Tree

```python
param_grid = {'max_depth':range(1,10),
             'max_leaf_nodes':range(1,10),
             'min_samples_leaf':range(1,10)} 
```

```python
grid_search = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5, return_train_score=True)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Optimal Parameters:", grid_search.best_params_)
print("Best Cross Validation Score: {:.2f}".format(grid_search.best_score_))
```

```python
df_results = pd.DataFrame(grid_search.cv_results_)
```

```python
df_results.drop(columns=['mean_fit_time', 'std_fit_time', 'mean_score_time', 'std_score_time', 'params'], inplace = True)
# 시간적인 부분을 버리고 필요한것만 보도록 해보자
```

```python
df_results = df_results.dropna(axis=0) # 유효하지 않은 파라미터(NaN) 제거
```

```python
df_results
```

```python
df_graph = df_results[['split0_test_score','split1_test_score','split2_test_score','split3_test_score','split4_test_score']]
df_graph.plot(alpha = 0.5)
```

데이터의 조합에 따라 수치가 분포가 다양하게 나뉘고 있음을 파악할 수 있다. 실제로 split 1의 경우 0.8을 넘는 높은 수치를 보이고 있으나, split 0과 2의 경우에는 낮은 수치를 보이는 것을 보아, 모델이 일반화가 잘 되지 않음을 파악할 수 있다. 하지만 평균적인 수치는 기존의 선형 모델에 비해 높은 정확도를 보이고 있다.
